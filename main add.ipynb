```python
import cv2
import numpy as np
input_dir = 'dataset/AddTest'
output_dir = 'dataset/AddOutput'
groundtruth_dir = 'dataset/AddTestGT'

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_no = '013'
```

```python
img = cv2.imread(input_dir + f'/SLA_{image_no}.jpg')
img_ground = cv2.imread(groundtruth_dir + f'/SLA_GT_{image_no}.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # convert BGR to RGB
plt.imshow(img_rgb)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Reduce noise
blur = cv2.GaussianBlur(img,(7,7),0)
img_contrast = cv2.convertScaleAbs(blur, alpha=1.2, beta=10)
```

```python
ret,binary = cv2.threshold(img_contrast,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
cross_kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
eroded = cv2.erode(binary,cross_kernel,iterations = 5)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121), plt.imshow(binary,cmap='gray')
plt.title('Binary Image')
plt.xticks([]),plt.yticks([])

plt.subplot(122), plt.imshow(eroded,cmap='gray')
plt.title('Eroded Image')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Flood fill to remove border image mask
thresh = eroded.copy()
h, w = img.shape[:2]
for row in range(h):
    if thresh[row, 0] == 255:
        cv2.floodFill(thresh, None, (0, row), 0)
    if thresh[row, w-1] == 255:
        cv2.floodFill(thresh, None, (w-1, row), 0)
for col in range(w):
    if thresh[0, col] == 255:
        cv2.floodFill(thresh, None, (col, 0), 0)
    if thresh[h-1, col] == 255:
        cv2.floodFill(thresh, None, (col, h-1), 0)

# Flood fill background to find inner holes
# holes = thresh.copy()
dilated = cv2.dilate(thresh,cross_kernel,iterations = 10)
holes = dilated.copy()
cv2.floodFill(holes, None, (0, 0), 255)

# invert holes thresh, bitwise or with thresh to fill in holes
holes = cv2.bitwise_not(holes)
filled_mask = cv2.bitwise_or(dilated, holes)
plt.imshow(filled_mask,cmap='gray')
plt.show()
```

```python
plt.imshow(dilated,cmap='gray')
plt.show()
```

```python
cnts, _ = cv2.findContours(filled_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cnt = max(cnts, key=cv2.contourArea)

out = np.zeros(filled_mask.shape, np.uint8)
cv2.drawContours(out, [cnt], -1, 255, cv2.FILLED)
ctr_mask = cv2.bitwise_and(filled_mask, out)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
dilation = cv2.dilate(ctr_mask,kernel,iterations = 5)
# plt.imshow(dilation,cmap='gray')
# plt.show()
```

```python
plt.figure(figsize=(8,15))
plt.subplot(421), plt.imshow(img,cmap='gray')
plt.title('Grayscaled Image')
plt.xticks([]),plt.yticks([])

plt.subplot(422), plt.imshow(blur,cmap='gray')
plt.title('After Gaussian Blur')
plt.xticks([]),plt.yticks([])

plt.subplot(423), plt.imshow(img_contrast,cmap='gray')
plt.title('Scaled Image')
plt.xticks([]),plt.yticks([])

plt.subplot(424), plt.imshow(binary,cmap='gray')
plt.title('Binary Image')
plt.xticks([]),plt.yticks([])

plt.subplot(425), plt.imshow(thresh,cmap='gray')
plt.title('Floodfill')
plt.xticks([]),plt.yticks([])

plt.subplot(426), plt.imshow(filled_mask,cmap='gray')
plt.title('Filled Holes')
plt.xticks([]),plt.yticks([])

plt.subplot(427), plt.imshow(ctr_mask,cmap='gray')
plt.title('Largest Area Extracted')
plt.xticks([]),plt.yticks([])

plt.subplot(428), plt.imshow(dilation,cmap='gray')
plt.title('Dilated Image')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121), plt.imshow(img_ground,cmap='gray')
plt.title('GT Image')
plt.xticks([]),plt.yticks([])

plt.subplot(122), plt.imshow(dilation,cmap='gray')
plt.title('Output Image')
plt.xticks([]),plt.yticks([])
plt.show()
```
